```python
import pandas
from datetime import date
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
data = pandas.read_csv("../Dataset/AlteredColumns.csv")
```

```python
data.head(5)
```

```python
data["PeriodStart"]=data["PeriodStart"].astype(str)
data["PeriodStart"]=data["PeriodStart"].str.replace('T',' ')
data["PeriodStart"]=data["PeriodStart"].str.replace('Z','')
data.head(5)
```

```python
data["PeriodStart"]=pandas.to_datetime(data["PeriodStart"], format='%Y-%m-%d %H:%M:%S')
data.head(5)
```

```python
data["hour"] = data["PeriodStart"].map(lambda x: x.hour)
data["minutes"] = data["PeriodStart"].map(lambda x: x.minute)
data["day"] = data['PeriodStart'].map(lambda x: x.day)
data["month"] = data['PeriodStart'].map(lambda x: x.month)
data["year"] = data['PeriodStart'].map(lambda x: x.year)
data = data.drop(['PeriodStart'], axis=1)
data.head(5)
```

```python
X = data.drop(["Ghi"], axis=1)
y = data["Ghi"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
acc = r2_score(y_test, y_pred)*100
```

```python
print("Accuracy of the model is: ", acc)
```
